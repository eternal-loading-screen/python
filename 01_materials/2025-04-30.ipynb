```python
# set a default value in function definition
def calc_sales_tax(price, tax_rate = 0.13 ):
    '''Convert Celsuis to Fahrenheit'''
    tax_amt = price * tax_rate
    return tax_amt
```

```python
# supplying less than 2 arguments
calc_sales_tax (5)
```

```python
# supply 2 arguments
calc_sales_tax(5, 0.08)
```

```python
def calc_total_bill(price, tax_rate = 0.13, tip_rate = 0.15):
    tax_amt = price * tax_rate
    tip_rate = price * tip_rate
    total_price = price + tax_amt + tip_rate
    return total_price
```

```python
calc_total_bill(100)
```

```python
calc_total_bill (100, tip_rate=  0.18)
```
